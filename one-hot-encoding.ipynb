```python
import pandas as pd
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
columns = data["whoAmI"].unique()
```

```python
one_hot_df = pd.DataFrame(columns=columns)
```

```python
one_hot_df.human = (data["whoAmI"] == 'human').replace({True: 1, False: 0})
one_hot_df.robot = (data["whoAmI"] == 'robot').replace({True: 1, False: 0})
one_hot_df
```
